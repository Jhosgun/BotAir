```python
# Librerias
from MakeSens import MakeSens
import pandas as pd 
import datetime
import pytz
import pylab as plt
import folium
from folium.plugins import BeautifyIcon
```

```python
# Rango de fecha
fechaFin = datetime.datetime.now(pytz.timezone('America/Bogota')).strftime('%Y-%m-%d %H:%M:%S')
fechaInicio = (datetime.datetime.now(pytz.timezone('America/Bogota'))-datetime.timedelta(hours=240)).strftime('%Y-%m-%d %H:%M:%S')
```

```python
# Coordenadas de Bucaramanga, Colombia
latitude = 7.11392
longitude = -73.1198

# Información de las estaciones
Estaciones = [ ('mE1_00004','Col. Tecnologico',7.1381572,-73.1180935),
               ('mE1_00005','Col. Santander',7.1385332,-73.121327),
               ('mE1_00007','Col.Normal',7.1277056,-73.116534),
               ('E2_00023','Grupo Halley',7.1393716,-73.1210968)]


# Crear el mapa de Bucaramanga con el estilo "CartoDB Positron"
bucaramanga_map = folium.Map(location=[latitude, longitude], 
                              zoom_start=20, tiles='CartoDB Positron')


def get_color(number):
    if 0 <= number <= 12:
        return '#00FF2E'  # Verde
    elif 13 <= number <= 37:
        return '#FFE000'  # Amarillo
    elif 38 <= number <= 55:
        return '#FF6100'  # Naranja
    elif 56 <= number:
        return '#FF0000'  # Rojo

    
for est in Estaciones:
    data = MakeSens.download_data(est[0], fechaInicio, fechaFin, 'd')
    data = data.sort_index(ascending=False)
    
    # Verificación del dato
    #print("Dato promedio ",est[1]," : ",data["pm25_1"][0])
    
    # Crear el ícono personalizado con un número
    icon_number = BeautifyIcon(
    border_color=get_color(data["pm25_1"][0]),
    text_color=get_color(data["pm25_1"][0]),
    number=round(data["pm25_1"][0],1),
    icon_size=(40, 40),  # Tamaño del ícono en píxeles
    inner_icon_style='line-height:25px; font-size: 15px;',  # Ajusta el tamaño de la fuente aquí
    spin=True,
    border_width=2,
    )

    # Agregar marcador del Halley
    folium.Marker(
        [est[2], est[3]],
        popup=est[1],
        icon=icon_number
    ).add_to(bucaramanga_map)
    
    
# Mostrar el mapa
bucaramanga_map
```
